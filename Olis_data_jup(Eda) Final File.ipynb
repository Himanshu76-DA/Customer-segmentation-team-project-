```python
import pandas as pd
STATUS_COL   = "order_status"                   
DELIVERED_DT = "order_delivered_customer_date"  

has_delivered_status = (
    STATUS_COL in df.columns and
    df[STATUS_COL].astype(str).str.lower() == "delivered"
)

has_delivered_date = (
    DELIVERED_DT in df.columns and
    df[DELIVERED_DT].notna()
)


if isinstance(has_delivered_status, pd.Series) and isinstance(has_delivered_date, pd.Series):
    delivered = has_delivered_status | has_delivered_date
elif isinstance(has_delivered_status, pd.Series):
    delivered = has_delivered_status
elif isinstance(has_delivered_date, pd.Series):
    delivered = has_delivered_date
else:
    raise ValueError("Neither STATUS_COL nor DELIVERED_DT exists in your DataFrame. Set the names at the top.")


total = len(df)
delivered_count = delivered.sum()
not_delivered_count = total - delivered_count
not_delivered_pct = (not_delivered_count / total) * 100

print(f"Total orders: {total}")
print(f"Delivered: {delivered_count}")
print(f"NOT delivered: {not_delivered_count}")
print(f"â— Percent NOT delivered: {not_delivered_pct:.2f}%")


if STATUS_COL in df.columns:
    print("\nOrders by status:")
    print(df[STATUS_COL].value_counts(dropna=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df["estimated_delivery_date"] = pd.to_datetime(df["estimated_delivery_date"], errors="coerce")
df["delivered_customer_date"] = pd.to_datetime(df["delivered_customer_date"], errors="coerce")
```

```python
def delivery_group(row):
    est = row["estimated_delivery_date"]
    act = row["delivered_customer_date"]

    if pd.isna(act):
        return "not_delivered"
    if pd.isna(est):
        return "unknown"
    return "on_time" if act <= est else "delayed"

df["delivery_group"] = df.apply(delivery_group, axis=1)
df["delivery_group"].value_counts()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def make_datetime(df, date_col, time_col=None):
    if time_col and time_col in df.columns and df[time_col].notna().any():
        if np.issubdtype(df[date_col].dtype, np.datetime64):
            date_str = df[date_col].dt.strftime("%Y-%m-%d")
        else:
            date_str = df[date_col].astype(str)
        dt = pd.to_datetime(date_str + " " + df[time_col].astype(str), errors="coerce")
    else:
        dt = pd.to_datetime(df[date_col], errors="coerce")
    return dt

ORDER_DATE_COL = "order date"
ORDER_TIME_COL = "order time"
APPROVED_DATE_COL = "approved date"
APPROVED_TIME_COL = "approved time"
DELIV_CUST_DATE_COL = "delivered_customer_date"
DELIV_CUST_TIME_COL = "delivered_customer_time"
STATE_COL = "customer_state"

df["order_dt"] = make_datetime(df, ORDER_DATE_COL, ORDER_TIME_COL)
df["approved_dt"] = make_datetime(df, APPROVED_DATE_COL, APPROVED_TIME_COL)
df["delivered_dt"] = make_datetime(df, DELIV_CUST_DATE_COL, DELIV_CUST_TIME_COL)

df["minutes_to_approved"] = (df["approved_dt"] - df["order_dt"]).dt.total_seconds() / 60
df["days_to_delivered"] = (df["delivered_dt"] - df["order_dt"]).dt.total_seconds() / (60*60*24)

state_speed = (
    df.groupby(STATE_COL)[["minutes_to_approved", "days_to_delivered"]]
      .mean()
      .round(2)
      .sort_values("days_to_delivered", ascending=False)
)

display(state_speed.head(10))
display(state_speed.tail(10))

slow15 = state_speed.head(15)
slow15["days_to_delivered"].plot(kind="barh", figsize=(9,5))
plt.gca().invert_yaxis()
plt.xlabel("Avg days from order to delivery")
plt.title("Slowest States (Avg Delivery Time)")
plt.tight_layout()
plt.show()
```

### Analyse the sudden rise in profit after 2017-10th month

```python
DATE_COLUMN = "order date"          
VALUE_COLUMN = "Profit"             
df[DATE_COLUMN] = pd.to_datetime(df[DATE_COLUMN], errors="coerce")
monthly_data = (
    df.dropna(subset=[DATE_COLUMN])
      .set_index(DATE_COLUMN)
      .resample("M")[VALUE_COLUMN]
      .sum()
)
plt.figure(figsize=(10,5))
plt.plot(monthly_data.index, monthly_data.values, marker='o', color='teal', linestyle='-')
plt.title(f"Monthly {VALUE_COLUMN} Over Time")
plt.xlabel("Month")
plt.ylabel(VALUE_COLUMN)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_before_hype = df[df["order date"] < '2017-11-30']
df_hype_data = df[(df["order date"] < '2018-01-01') & (df["order date"] > '2017-11-30')]
```

```python
DATE_COLUMN = "order date"
VALUE_COLUMN = "Profit"

df_before_hype[DATE_COLUMN] = pd.to_datetime(df_before_hype[DATE_COLUMN], errors="coerce")
monthly_data = (
    df_before_hype.dropna(subset=[DATE_COLUMN])
    .set_index(DATE_COLUMN)
    .groupby('customer_state')[VALUE_COLUMN]
    .resample('M')
    .sum()
    .reset_index()
)

plt.figure(figsize=(14,6))
sns.lineplot(data=monthly_data, x=DATE_COLUMN, y=VALUE_COLUMN, hue='customer_state', linewidth=2.2)
plt.title(f"Monthly {VALUE_COLUMN} Trend by Customer State", fontsize=18)
plt.xlabel("Month", fontsize=14)
plt.ylabel(VALUE_COLUMN, fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Customer State', bbox_to_anchor=(1.02, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df_hype_data = df_hype_data.groupby('customer_state')['Profit'].sum().reset_index().sort_values('Profit', ascending=False)

plt.figure(figsize=(12,6))
sns.barplot(data=df_hype_data, x='customer_state', y='Profit', palette='viridis')
plt.title('Total Profit by Customer State', fontsize=18)
plt.xlabel('Customer State', fontsize=14)
plt.ylabel('Total Profit', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
 
```
