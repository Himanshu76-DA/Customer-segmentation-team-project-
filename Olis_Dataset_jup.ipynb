### Since our dataset contains only unique customers with no repeat purchases, we used the total number of items bought (from the `order_item_id` column) as a proxy for Frequency in the RFM model. This approach allows us to retain the traditional RFM structure while still capturing customer purchase intensity.


### There is only one extreme outlier have Profit around 12000, Lets seperate it 

```python
df = df[df["Profit"]  < 6000]
```

```python
df_p = df[df["Profit"] > 0]
# df_p conatins data that have profit > 0
```

```python
df.rename(columns = {"order_item_id" : "items"},inplace = True)
```

```python
df["items"] = df["items"].astype('int')
```

```python
RFM = pd.DataFrame({"customer_id" : df["customer_id"],
                    "Recency" : df["order date"].max() - df["order date"]})
RFM["Frequency"] = df.groupby("customer_id")["items"].sum().values
RFM["Monetary"] = df.groupby("customer_id")["Profit"].sum().values
```

```python
RFM.info()
```

```python
RFM["Recency"] = RFM["Recency"].astype('int')
```

```python
std_col = RFM[['Recency', 'Frequency', 'Monetary']]

from sklearn.preprocessing import StandardScaler
stded_col = StandardScaler().fit_transform(std_col)

from sklearn.cluster import KMeans
RFM["Cluster"] = KMeans(n_clusters = 4, random_state = 10).fit_predict(stded_col)

sns.scatterplot(data = RFM, x = "Monetary", y = "Recency", hue = "Cluster", palette = "viridis")
```

```python
RFM["Recency"] = (pd.to_datetime(df["order date"].max()) - pd.to_datetime(df["order date"])).dt.days
```

### Lets check the best cluster size

```python
data = RFM[['Recency', 'Frequency', 'Monetary']] 
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=10)
    kmeans.fit(stded_col)
    wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.xticks(range(1,11))
plt.show()
```
